This corpus has been collected from free or free for research sources at the Internet:

-> A collection of 425 SMS spam messages was manually extracted from the Grumbletext Web site. This is a UK forum in which cell phone users make public claims about SMS spam messages, most of them without reporting the very spam message received. The identification of the text of spam messages in the claims is a very hard and time-consuming task, and it involved carefully scanning hundreds of web pages. The Grumbletext Web site is: [Web Link].

<b> Data Processing

```
import re
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```
# initialise the inbuilt stemmer
stemmer= PorterStemmer()
```

```
# We can also use Lemmatizer instead of stemmer
Lemmatizer=WordNetLemmatizer()
```

```
# initalizing the lists
lst1=[]
lst1=[]
```

```
def preprocess(raw_data,  flag):  


  # Removing special character and digits
  letters_only  = re.sub("[^a-zA-Z]"," ",raw_data) 

  # change sentence to lower case
  letters_only  = letters_only.lower() 

  # tokenize into words
  words = letters_only.split()  

  #remove stop words
  words = [w for w in words if words if not w in stopwords.words("english")]              

  #stemming/lemmatization
  if(flag ==  'stem'):
      words = [stemmer.stem(word) for word in words]
  else:
      words =  [Lemmatizer. lemmatize(word) for word in words]

  lst1.append(" ".join(words))
  lst2.append(len(words))
```

```
 import nltk
 nltk.download('stopwords')
lst1 = []
lst2 = []
df['message'].apply(lambda x: preprocess(x,'stem'))
df['clean_message_stem'] = lst1
df['message_length_steam'] = lst2
```

```

lst1=[]
lst2=[]
df['message'].apply(lambda x: preprocess(x,'lemma'))
df['clean_mesaage_lema']=lst1
df['message_length_lema']=lst2
```

<b> Step-4 Data Preparation

```
# Train Test Split
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2,random_state=42)
```

```
train_clean_message=[]
for message1 in train['clean_message_stem']:
  train_clean_message.append(message1)

  test_clean_message=[]
for message1 in train['clean_message_stem']:
  test_clean_message.append(message1)

```

<b> Text to numerical vector using Bag of Words 

```
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer(analyzer="word")
train_features=vectorizer.fit_transform(train_clean_message)
test_features=vectorizer.transform(test_clean_message)
```

```
vectorizer.vocabulary_
```

```
print("Total unique words:",len(vectorizer.vocabulary_))
print("Type of Train_feature:",type(train_features))
print("Shape of input data",train_features.shape)
```

<b> <b>Step-5 Train the model. Use the following algorithm

```
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
from sklearn import metrics
```

```
Classifier=[
            LogisticRegression(),
            DecisionTreeClassifier(),
            SVC()]
```

<b> Step-6 Evaluation for model ,evaluate using following metrics:

-Check accuracy score

-Build confusion metric

-Print Classification_report

```
from tqdm import tqdm, tqdm_notebook

tqdm.pandas()
```

```
#check accuracy score
dense_features = train_features.toarray()
dense_test = test_features.toarray()
for classifier in tqdm(Classifier):
  fit=classifier.fit(dense_features,train['target_new'])
  pred=fit.predict(dense_test)
  accuracy = accuracy_score(pred,test['target_new'])
  print('Accuracy of '+classifier.__class__.__name__+'is'+str(accuracy))
   
   # confusion metric
  con_metric=metrics.confusion_matrix(test['target_new'],pred)
  print(con_metric)

  # Print Classification_report
  print(metrics.classification_report(test['target_new'],pred))
   
```
