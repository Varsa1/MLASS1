<b> Missing values handling

```
miss=['?']
new_df=pd.read_csv('adult.csv',na_values=miss)
new_df
```

```
new_df.isnull()
```

```
new_df.isnull().sum()
```

```
new_df.isnull().any(axis=0)
```

```
round(100*(new_df.isnull().sum()/len(new_df.index)),2)
```

```
# remove rows which having missing values
new_df.dropna(axis=0,how='any',inplace=True)

# remove 'fnlwgt' column as it bis not required
new_df.drop('fnlwgt',axis=1,inplace=True)
new_df.shape
print('rows',new_df.shape[0])
print('columns',new_df.shape[1])
```

```
# percentage of rows with missing values
missingvalues=((df.shape[0]-new_df.shape[0])/df.shape[0])
print(missingvalues)
```
